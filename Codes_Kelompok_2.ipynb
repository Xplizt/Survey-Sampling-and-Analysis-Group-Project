```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Survey Pengaruh Gaya Hidup dan Online Shop Terhadap Perilaku Konsumtif Mahasiswa (Responses) - Raw Data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Timestamp'], inplace=True)
df.drop(columns=['Nama'], inplace=True)
```

```python
column_data=df['Apakah Anda pernah berbelanja online?'].value_counts()
colors = ['#355e3b', '#3a7f41', '#4e9258', '#66cdaa']

plt.figure(figsize=(4, 4))
plt.pie(column_data, labels=column_data.index, autopct='%1.1f%%', colors=colors)
plt.title('Apakah Anda pernah berbelanja online?')
plt.show()
```

```python
threshold = len(df) * 0.5
df.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
df.head()
```

```python
num_cols = len(df.select_dtypes(include=['object']).columns)
num_rows = 2
num_cols_per_row = 3

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols_per_row, figsize=(num_cols_per_row * 8, num_rows * 6))

for i, column in enumerate(df.select_dtypes(include=['object']).columns):
    row = i // num_cols_per_row
    col = i % num_cols_per_row
    df[column].value_counts().plot(kind='bar', color='green', ax=axes[row, col])
    axes[row, col].set_title(f'Bar Chart of {column}')
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
summary_stats = df.describe()
```

```python
green_palette = sns.light_palette("green", as_cmap=True)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(summary_stats, annot=True, cmap=green_palette, linewidths=.5)
plt.title('Summary Statistics Heatmap')
plt.show()
```

```python
mean_values = df.mean(numeric_only=True)
plt.figure(figsize=(10, 6))
mean_values.plot(kind='bar', color=plt.cm.Greens(np.linspace(0, 1, len(mean_values))))
plt.title('Mean Values of Each Column')
plt.ylabel('Mean')
plt.xlabel('Column')
plt.show()
```

```python
std_dev = df.std(numeric_only=True)
plt.figure(figsize=(10, 6))
std_dev.plot(kind='bar', color=plt.cm.Greens(np.linspace(0, 1, len(std_dev))))
plt.title('Standard Deviation of Each Column')
plt.ylabel('Standard Deviation')
plt.xlabel('Column')
plt.show()
```

```python
variance = df.var(numeric_only=True)
plt.figure(figsize=(10, 6))
variance.plot(kind='bar', color=plt.cm.Greens(np.linspace(0, 1, len(variance))))
plt.title('Variance of Each Column')
plt.ylabel('Variance')
plt.xlabel('Column')
plt.show()
```

```python
for column in df.select_dtypes(include=['object']).columns:
    freq_count = df[column].value_counts()
    plt.figure(figsize=(10, 6))
    freq_count.plot(kind='bar', color=plt.cm.Greens(np.linspace(0, 1, len(freq_count))))
    plt.title(f'Frequency Count for Column {column}')
    plt.ylabel('Count')
    plt.xlabel('Category')
    plt.show()
```

```python
df2 = pd.read_csv('Survey Pengaruh Gaya Hidup dan Online Shop Terhadap Perilaku Konsumtif Mahasiswa (Responses) - CLEANED.csv')
```

```python
# Plotting
fig, axs = plt.subplots(9, 1, figsize=(10,10))

# Flatten axes for easy iteration
axs = axs.flatten()

columns_to_plot = [
    'Seberapa sering kamu berbelanja secara online?',
    'Seberapa sering kamu merasa tergoda untuk membeli barang saat melihat postingan atau iklan online?',
    'Seberapa sering kamu membeli barang yang sebenarnya tidak Anda butuhkan?',
    'Seberapa sering kamu membandingkan diri dengan orang lain berdasarkan barang yang dimiliki mereka?',
    'Seberapa sering kamu menggunakan aplikasi online shop sebagai penghibur saat merasa bosan atau stres?',
    'Seberapa sering kamu memperhatikan merek atau label saat berbelanja online?',
    'Seberapa sering kamu membeli barang online dengan harga yang melebihi anggaran belanja bulanan kamu?',
    'Seberapa sering kamu merasa puas setelah melakukan pembelian online?',
    'Apakah kamu sering merasa cemas atau stres terkait dengan pengeluaran bulanan kamu untuk belanja online?'
]

for i, column in enumerate(columns_to_plot):
    ax = axs[i]
    values = df2[column]
    ax.bar(values.unique(), values.value_counts(), color='skyblue', width=0.5)
    ax.set_title(column)
    ax.set_xlabel('Rating')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# List of columns to be considered
columns_to_include = ['Shopee', 'Tokopedia', 'TikTok Shop', 'Zalora', 'Bukalapak', 'Lazada', 'OLX', 'Kaskus FJB', 'Blibli']

# Select only the specified columns
df_selected = df2[columns_to_include]

# Ensure all selected columns are of numeric type
df_selected = df_selected.apply(pd.to_numeric)

# Sum the values in each one-hot encoded column
counts = df_selected.sum()

# Calculate the total number of entries (sum of all counts)
total_count = counts.sum()

# Calculate the percentage for each category
percentages = (counts / total_count) * 100

# Data for the pie chart
labels = percentages.index
sizes = percentages.values
colors = plt.cm.Greens(np.linspace(0.2, 0.8, len(labels)))

# Plotting the pie chart
plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

plt.title('Percentage of Entries by Platform')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the plot
plt.show()
```
